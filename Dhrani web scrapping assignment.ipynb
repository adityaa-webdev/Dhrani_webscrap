```python
#importing the required libraries
from selenium import webdriver
from selenium.webdriver.support.ui import Select
from selenium.webdriver.common.by import By
```

```python
#create a chrome webdriver instance to automate the navigation process
driver = webdriver.Chrome()
driver.get("https://dharani.telangana.gov.in/knowLandStatus")
```

```python
# Select district, mandal and village using their respective dropdown IDs or names
district_dropdown = Select(driver.find_element(By.ID, 'districtID'))  
mandal_dropdown= Select(driver.find_element(By.ID,"mandalID"))
village_dropdown = Select(driver.find_element(By.ID,"villageId"))
```

```python
##create a helper function which takes in the desried district_Name,mandal_name,village_name and returns a list of survey_numbers
def get_survey_numbers(DistrictName,MandalName,VillageName):
    district_dropdown.select_by_visible_text(DistrictName)
    mandal_dropdown.select_by_visible_text(MandalName)
    village_dropdown.select_by_visible_text(VillageName)
    survey_id = Select(driver.find_element(By.ID,"surveyIdselect"))
    options = survey_id.options
    survey_id_list = []

    for option in options[1:]:
        survey_id_list.append(option.text) 
    return survey_id_list
```

```python
result=get_survey_numbers("Adilabad|ఆదిలాబాద్","Adilabad (Rural)|ఆదిలాబాద్ (రూరల్)","Ankapoor|అంకాపూర్")
result
```
